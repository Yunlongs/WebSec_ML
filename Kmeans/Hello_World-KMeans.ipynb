### 10.2 示例：Hello World！ KMeans

这里面需要用到sklearn的聚类用的数据集make_blobs.

`sklearn.datasets.make_blobs(n_samples=100, n_features=2, centers=None, cluster_std=1.0, center_box=(-10.0, 10.0), shuffle=True, random_state=None)`

>https://scikit-learn.org/stable/modules/generated/sklearn.datasets.make_blobs.html#sklearn.datasets.make_blobs

```python
import numpy as np
import matplotlib.pyplot as plt
from sklearn.cluster import KMeans
from sklearn.datasets import make_blobs

n_samples = 1500
x,y = make_blobs(n_samples,random_state=170)

y_pred = KMeans(n_clusters=3,random_state=170).fit_predict(x)

plt.scatter(x[:,0],x[:,1],c=y_pred)
plt.title("hello world!")
plt.show()
```

#### KMeans介绍

`class sklearn.cluster.KMeans(n_clusters=8, init=’k-means++’, n_init=10, max_iter=300, tol=0.0001, precompute_distances=’auto’, verbose=0, random_state=None, copy_x=True, n_jobs=None, algorithm=’auto’)`
**下面来介绍下KMeans的参数**：

- n_clusters:聚类数
- init：初始化的方法。‘k-means++’：选取距离较远的初始聚类，加速收敛。‘random’：随机选择。
- n_init：使用不同的聚类中心运行几次算法
- max_iter：最大迭代数
- tol：收敛的判断参数
- precompute_distances：提前计算距离（速度更快，但更消耗内存）。‘auto’ ：当样本过大时不使用。'True'：是。‘False’：不使用。
- copy_x：不太懂
- algorithm：'full'：经典的EM类算法。‘elkan’：更高效但不支持稀疏数据。'auto'：数据稀疏时使用full，不稀疏时使用elkan。

>https://scikit-learn.org/stable/modules/generated/sklearn.cluster.KMeans.html#sklearn.cluster.KMeans
